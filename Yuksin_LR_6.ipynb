```python
import altair as alt
from vega_datasets import data
import pandas as pd

weather = data.seattle_weather()

alt.Chart(weather).mark_bar(
    cornerRadiusTopLeft=10,
    cornerRadiusTopRight=10
).encode(
    x=alt.X('weather:N', sort='-y', title=' '),
    y=alt.Y('count():Q', title=' '),
    color=alt.Color('weather:N', title='')
).properties(title='    ')
```

```python
barley = data.barley()

alt.Chart(barley).transform_filter(
    alt.datum.year == 1931
).mark_bar().encode(
    x=alt.X('yield:Q', title=''),
    y=alt.Y('variety:N', sort='-x', title=''),
    color=alt.Color('site:N', title=' ')
).properties(title='    (1931)')
```

```python
alt.Chart(weather).mark_bar().encode(
    x=alt.X('month(date):N', title=''),
    y=alt.Y('count():Q', title=''),
    color=alt.Color('weather:N', title=' '),
    xOffset='weather:N'
).properties(title='   ')
```

```python
stocks = data.stocks()

base = alt.Chart(stocks).transform_filter(
    alt.FieldOneOfPredicate(field='symbol', oneOf=['GOOG', 'AAPL'])
).encode(
    x='date:T',
    y='price:Q',
    color=alt.Color('symbol:N')
)

area = base.mark_area(
    line={'color':'darkblue'},
    color=alt.Gradient(
        gradient='linear',
        stops=[alt.GradientStop(color='white', offset=0),
               alt.GradientStop(color='blue', offset=1)],
        x1=1, x2=1, y1=1, y2=0
    )
)

area.properties(title='   GOOG  AAPL')
```

```python
unemp = data.unemployment_across_industries.url

selection = alt.selection_point(fields=['series'], bind='legend')

alt.Chart(unemp).mark_area().encode(
    alt.X('date:T').axis(format='%Y'),
    alt.Y('sum(count):Q').stack('center'),
    alt.Color('series:N').scale(scheme='category20'),
    opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
).add_params(
    selection
).interactive().properties(width=600, title='   ')
```

```python
weather_data = data.seattle_weather()
avg_temp = weather_data['temp_max'].mean()

base = alt.Chart(weather_data).encode(
    theta=alt.Theta("count():Q", stack=True),
    color=alt.Color("weather:N")
)

pie = base.mark_arc(innerRadius=50)

text = alt.Chart(pd.DataFrame({'avg': [avg_temp]})).mark_text(
    size=20, color='black'
).encode(text=alt.Text('avg:Q', format='.1f'))

(pie + text).properties(title='    . . ')
```
