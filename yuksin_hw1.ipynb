```python
import numpy as np
# 1. Створіть одновимірний масив (вектор) з першими 20-ма натуральними числами та виведіть його значення.
vector = np.arange(1, 21)
print("Вектор (1-20):\n", vector)
```

```python
# 2. Створіть двовимірний масив (матрицю) розміром 4x4, заповніть його одиницями та виведіть його значення.
matrix_1=np.ones((4, 4))
print("Матриця 4х4 з одиниць:\n", matrix_1)
```

```python
# 3. Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від -1 до 1 та виведіть його значення.
matrix_2 = np.random.uniform(-1, 1, (5, 5))
print("Матриця 5x5 (-1 до 1):\n", matrix_2)
```

```python
# 4. Створіть масив розміром 3x3, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та виведіть його значення.
matrix_3 = np.random.rand(3, 3)
print("Матриця 3x3 (0 до 1):\n", matrix_3)
```

```python
# 5. Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими числами в діапазоні від 1 до 20
# та виконайте на них поелементні операції додавання, віднімання та множення.
arr1 = np.random.randint(1, 21, 6)
arr2 = np.random.randint(1, 21, 6)

print("Масив 1:", arr1)
print("Масив 2:", arr2)
print("Додавання:", arr1 + arr2)
print("Віднімання:", arr1 - arr2)
print("Множення:", arr1 * arr2)
```

```python
# 6. Створіть дві матриці розміром 3x2 та 2x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.
mat1 = np.random.randint(1, 6, (3, 2))
mat2 = np.random.randint(1, 6, (2, 4))

product_matrix = np.dot(mat1, mat2) # Або mat1 @ mat2

print("Матриця 3x2:\n", mat1)
print("Матриця 2x4:\n", mat2)
print("Результат множення (3x4):\n", product_matrix)
```

```python
# 7. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.
m_orig = np.random.rand(3, 3)
m_transposed = m_orig.T

print("Оригінальна матриця:\n", m_orig)
print("Транспонована матриця:\n", m_transposed)
```

```python
# 8. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню різницю.
m_a = np.random.randint(1, 11, (3, 3))
m_b = np.random.randint(1, 11, (3, 3))

print("Матриця A:\n", m_a)
print("Матриця B:\n", m_b)
print("Різниця (A - B):\n", m_a - m_b)
```

```python
# 9. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток.
# Примітка: для матричного множення використовуємо np.dot або @, для поелементного - *
m_c = np.random.randint(-5, 6, (3, 3))
m_d = np.random.randint(-5, 6, (3, 3))

matrix_prod = np.dot(m_c, m_d)

print("Матриця C:\n", m_c)
print("Матриця D:\n", m_d)
print("Матричний добуток:\n", matrix_prod)
```

```python
# 10. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1
# та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.
m_last = np.random.uniform(-1, 1, (3, 3))
row_sums = np.sum(m_last, axis=1).reshape(-1, 1) # reshape робить з цього вектор-стовпчик

print("Матриця:\n", m_last)
print("Вектор-стовпчик сум рядків:\n", row_sums)
```
